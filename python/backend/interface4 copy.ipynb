```python
paths = ["./fot10_copie.xlsx","./fot11_copie.xlsx"]

# La on veut comparer le rapport 11 au rapport 10. On a 3 cas : 
# cas 1 : fot10 conformes fot11 non conformes
# cas 2 : fot10 non conforme fot11 non conforme
# cas 3 : fot10 non conforme fot11 conforme
# Si on récupère liste des non conformes du 11. On a par exemple 
# fot11 = [19,101,104,108,112,200,391] On recupère la liste des non conformes du 10 On a par exemple
# fot10 = [19,65,104,108,150,200,391,450]
# pour chaque valeur de fot10 on regarde toute la liste de fot11. 
#           S'il est dedans : cas2 ici --> 19, 104, 200, 391
#           Elseif n'est pas dedans : cas 3 ici --> 65, 150, 450
#           le reste de fot11 : cas 1 ici --> 101, 112


liste_dic = []
for p in paths:
    cotes_n = {}
    df = select_file.select_file(p)
    dc = clean_data.clean_data(df)
    liste_dic.append(dc)

# Compter les occurrences dans chaque DataFrame
counts = [df['N° Cote'].value_counts() for df in liste_dic]

# Initialiser un dictionnaire pour stocker les indices à supprimer pour chaque DataFrame
indices_to_drop = {0: [], 1: []}

for value in set(counts[0].index).union(set(counts[1].index)):
    count_1 = counts[0].get(value, 0)
    count_2 = counts[1].get(value, 0)
    
    # Calculer la différence et déterminer de quel DataFrame supprimer les lignes
    if count_1 > count_2:
        diff = count_1 - count_2
        indices_to_drop_df = 0
    elif count_2 > count_1:
        diff = count_2 - count_1
        indices_to_drop_df = 1
    else:
        continue  # Si égal, passer à la valeur suivante

    # Obtenir les indices pour la valeur spécifique
    indices_for_value = liste_dic[indices_to_drop_df][liste_dic[indices_to_drop_df]['N° Cote'] == value].index

    # Sélectionner les indices à supprimer
    for i in range(diff):
        idx = indices_for_value[i]
        indices_to_drop[indices_to_drop_df].append(idx)
        next_idx = idx + 1
        if next_idx in liste_dic[indices_to_drop_df].index:
            indices_to_drop[indices_to_drop_df].append(next_idx)

# Appliquer les suppressions
for df_index in indices_to_drop:
    liste_dic[df_index].drop(indices_to_drop[df_index], inplace=True)






"""new_list = []
for dataframe in liste_dic:
    cotes_n = {}
    ds = result.filter_conformity(dataframe)
    df_filtered = ds[1][ds[1]['N° Cote'] > 10]
    for index, row in df_filtered.iterrows():
        cotes_n[index] = row['N° Cote'] 
    new_list.append(cotes_n)"""

#dict1 = new_list[0]
#dict2 = new_list[1]
#print(dict1)
#print(dict2)
pd.concat([liste_dic[0]["N° Cote"],liste_dic[1]["N° Cote"]], axis=1)
```
